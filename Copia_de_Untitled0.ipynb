```python
import random
from deap import base, creator, tools, algorithms

# Definir la función de evaluación
def evaluate(individual):
    return abs(sum(individual) - 0),

# Configuración de DEAP
creator.create("FitnessMin", base.Fitness, weights=(-1.0,))
creator.create("Individual", list, fitness=creator.FitnessMin)

toolbox = base.Toolbox()
toolbox.register("attr_bool", random.randint, 0, 1)
toolbox.register("individual", tools.initRepeat, creator.Individual, toolbox.attr_bool, n=10)
toolbox.register("population", tools.initRepeat, list, toolbox.individual)

toolbox.register("evaluate", evaluate)
toolbox.register("mate", tools.cxTwoPoint)
toolbox.register("mutate", tools.mutFlipBit, indpb=0.05)  # Aumentar la probabilidad de mutación
toolbox.register("select", tools.selTournament, tournsize=3)

def main():
    population = toolbox.population(n=200)
    found_solution = False
    num=0
    for gen in range(50):
        print(f"Generación {gen+1}")
        num=0
        for ind in population:
            num=num+1
            print(f"Individuo #{num}: ",ind)

            if sum(ind) == 0:
                print("¡Solución encontrada!")
                print("Individuo:", ind)
                found_solution = True
                break

        if found_solution:
            break

        algorithms.eaSimple(population, toolbox, cxpb=0.5, mutpb=0.2, ngen=1, verbose=False)

if __name__ == "__main__":
    main()
```
