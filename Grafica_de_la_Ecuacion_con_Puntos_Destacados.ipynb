Aqui se hace lo que es la parte del  DEAP, el fitnessMin hace minimizar la aptitud, el individual hace una lista y tiene como propiedad fitnessMin,  los toolbox registra las operaciones etc, el tama√±o de la poblacion es de 100 individuos y con el best_individual se van a almacenar en lo que son las variables, a, b, c, d y e

```python
# DEAP
creator.create("FitnessMin", base.Fitness, weights=(-1.0,))
creator.create("Individual", list, fitness=creator.FitnessMin)

toolbox = base.Toolbox()
toolbox.register("attr_float", random.uniform, -10, 10)
toolbox.register("individual", tools.initRepeat, creator.Individual, toolbox.attr_float, n=5)
toolbox.register("population", tools.initRepeat, list, toolbox.individual)

# Create a population
population_size = 100
population = toolbox.population(n=population_size)

# Best individual
best_individual = tools.selBest(population, k=1)[0]
a, b, c, d, e = best_individual

```
