```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import MinMaxScaler
import tensorflow as tf
from tensorflow import keras
```

```python
#Leemos el archivo
data = pd.read_csv("bdop.csv")

#Extraemos la columna que se va a predecir
X = data.drop(columns=['pH_CAMPO'])
y = data['pH_CAMPO']

#La normalizamos usando un scaler
scaler = MinMaxScaler()
X_scaled = scaler.fit_transform(X)

#La dividimos en datos de entrenamiento y datos de prueba 80/20
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.2, random_state=42)




```

```python
# Creamos una red neuronal simple, podemos modificarla para mejores resultados
model = keras.Sequential([
    keras.layers.Input(shape=(X_train.shape[1],)),  # Input layer with the number of features
    keras.layers.Dense(64, activation='relu'),      # Hidden layer with 64 neurons and ReLU activation
    keras.layers.Dense(32, activation='relu'),      # Hidden layer with 32 neurons and ReLU activation
    keras.layers.Dense(1, activation='linear')      # Output layer for regression
])

# Compilamos el modelo
model.compile(optimizer='adam', loss='mean_squared_error')

# Entrenamos el model
model.fit(X_train, y_train, epochs=100, batch_size=32, validation_split=0.2)
```

```python
# Evaluamos el modelo para calcular el error medio cuadrado
mse = model.evaluate(X_test, y_test)
print(f"Error medio cuadrado: {mse}")
```

```python
# Colocamos la informacion a la que queremos predecir
new_data_row = np.array([[24196,	2.356,	2.35,	0.003,	0.0341,	19.6195,	0,	19.6566,	0.003,	0.208	,0.141,	20,	0.009,	768,	76.6667,	46,	35.3,	24.6,	1200,	83.7,	5.26]])

# Normalizamos
new_data_row_scaled = scaler.transform(new_data_row)

# Hacemos la prediccion
prediction = model.predict(new_data_row_scaled)

```
